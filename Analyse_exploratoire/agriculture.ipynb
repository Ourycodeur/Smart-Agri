* Conclusion : 
    * Le dataset crop_yield.csv a été écarté de l'étude pour deux raisons principales. Premièrement, il contient des rendements agricoles négatifs (Yield_tons_per_hectare < 0), ce qui n'est pas physiquement plausible pour une production agricole. Deuxièmement, la variable géographique est représentée par des régions génériques (North, South, East, West) plutôt que par des pays réels, ce qui est incompatible avec l'objectif du projet basé sur des données par pays (Area) et par année. Enfin, sa structure ne contient pas certaines variables essentielles utilisées dans le modèle final, notamment l'année et l'utilisation de pesticides. Le dataset a donc été jugé non pertinent pour répondre aux objectifs de prédiction et de recommandation agricole du projet.
